## Load the data
1. Read the IGN data file into a dataframe named `ign_scores`.  
2. Use the `"Platform"` column to label the rows.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

IGN="https://raw.githubusercontent.com/csbfx/advpy122-data/master/ign_scores.csv"
## Your code here . . .
# Read the IGN data file into a df ign_scores
ign_scores = pd.read_csv(IGN)

# Have "Platform" be the index and then reset it back to a column
ign_scores.set_index("Platform", inplace=True)
ign_scores.reset_index(inplace=True)

# Check to make sure the data looks correct
ign_scores.head()
```

## Problem 5
Use the Pokemon dataset to create a clustermap with color. First, filter the dataframe to only keep data with `Type 1` equals to following values: `Water`, `Normal`, `Grass`, `Bug` and `Psychic`. Annotate the dendrogram using different colors for these five different `Type 1` values. Use `Name` as the index.

pokemon_data is in https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv

```python
# Put the data into a df
pokemon_data = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv"
pkmn = pd.read_csv(pokemon_data)

# Check how it is organized
pkmn.head()
```

```python
# Problem 5
# Your code here . . .

# Filter the df to keep only Type 1 = Water, Normal, Grass, Bug, or Psychic
type_filter = ["Water", "Normal", "Grass", "Bug", "Psychic"]
pkmn_filtered = pkmn[pkmn["Type 1"].isin(type_filter)].copy()

# Use Pokemon names as row labels
pkmn_filtered.set_index("Name", inplace=True)

# Add colors for each Type 1
colors = {
    "Water" : "blue",
    "Normal" : "gray",
    "Grass" : "green",
    "Bug" : "olive",
    "Psychic" : "purple"
}
row_colors = pkmn_filtered["Type 1"].map(colors)

# Keep only numeric columns for clustering
pkmn_numeric = pkmn_filtered[[
    "Total", "HP", "Attack", "Defense",
    "Sp. Atk", "Sp. Def", "Speed", "Generation"
]]

# Create the cluster map
sns.clustermap(
    pkmn_numeric,
    row_colors=row_colors,
    figsize=(12,12),
)

plt.show()

```
