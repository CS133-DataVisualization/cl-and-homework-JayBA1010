### Question 1

Read in the data from the csv file above. Skip all the comment lines, create a header for the dataframe based on the fields list in the comment line that starts with **# Fields:**. Drop the first column `query acc.ver`. Set the `subject acc.ver` as the index of the dataframe. The last column of the dataframe should be `publications`. Show the first five rows of the resulting dataframe.

```python
# Imports
import pandas as pd
import urllib.request
blast_results = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/blast_results.csv"
blast = pd.read_csv(blast_results)
```

```python
### Look at the first few lines 
### We see that the data only starts to appear after index 8
with urllib.request.urlopen(blast_results) as response:
    for i in range(15):
        line = response.readline().decode('utf-8').strip()
        print(f"{i + 1}: {line}")
```

```python
### Your code here . . .
my_columns = ["query acc.ver", "subject acc.ver", "% identity", "alignment length", "mismatches",
              "gap opens", "q. start", "q. end", "s. start", "s. end", "evalue", "bit score",
              "% positives", "bonds", "publications"]
blast = pd.read_csv(blast_results,
                    skiprows=8,
                    header=None,
                    names=my_columns,
                    usecols=my_columns,
                    thousands="," # Mainly for Q3 where commas are used in bonds
                    )
blast.head()
```

### Question 2

### Question 3
List the `subject acc.ver` that has over 15,000 bonds.

```python
### Your code here . . .
# Filter the df to take rows where bonds > 15000 and only display the subject acc.ver and bonds columns
over_15000 = blast[blast["bonds"] > 15000][["subject acc.ver", "bonds"]]
over_15000.head()
```

### Question 4
Create a plot that shows the correlation between `% identity` and `% positives`.

```python
### Your code here . . .
# Import seaborn and matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(
    data=blast,
    x="% identity",
    y="% positives",
)
plt.title("% positives vs % identity\nHigher % identity results in higher % positives") 
```
